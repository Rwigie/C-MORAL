# Eval logs: iteration 0 vs max reward_mean
Provide a JSONL path, extract iteration 0 and the row with the maximum reward_mean, then display a table.

```python
import json
from pathlib import Path
import pandas as pd
from props.props_info import MOLO_PROPERTIES

def _infer_prop_key(col):
    if col in MOLO_PROPERTIES:
        return col
    if col.endswith('_value'):
        key = col[:-6]
        if key in MOLO_PROPERTIES:
            return key
    return None


def _direction_arrow(prop_key):
    info = MOLO_PROPERTIES.get(prop_key)
    if not info:
        return ''
    if info.optimization_direction == 'maximize':
        return '↑'
    if info.optimization_direction == 'minimize':
        return '↓'
    return ''


def show_eval_log_summary(log_path):
    log_path = Path(log_path)

    iter0 = None
    max_item = None
    max_reward = None

    with log_path.open() as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            try:
                obj = json.loads(line)
            except Exception:
                continue
            if obj.get('iteration') == 0:
                iter0 = obj
            rm = obj.get('reward_mean')
            if isinstance(rm, (int, float)):
                if max_reward is None or rm > max_reward:
                    max_reward = rm
                    max_item = obj

    rows = [iter0, max_item]
    labels = ['Baseline', 'RL Best']

    df = pd.DataFrame(rows, index=labels)
    # Drop score-related columns
    df = df[[c for c in df.columns if 'score' not in c]]

    # Append direction arrows to property columns based on MOLO_PROPERTIES
    new_columns = []
    for col in df.columns:
        key = _infer_prop_key(col)
        if key:
            arrow = _direction_arrow(key)
            if arrow:
                new_columns.append(f"{col} {arrow}")
                continue
        new_columns.append(col)
    df.columns = new_columns

    return df
```

```python
# Set this to any eval_logs.jsonl you want to analyze
bpq_log_path = 'runs/bpq/20260114_185913/eval_logs.jsonl'
print("=====================Evaluation Log Summary: BPQ=====================")
print(show_eval_log_summary(bpq_log_path))
```
