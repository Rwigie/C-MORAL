```python
from rdkit import Chem
from rdkit.Chem import Draw

# 1. 输入你的 SMILES 字符串
smiles = "CC1CCN(C(=O)CNC(=O)c2cc(OC)c(OC)c(OC)c2)C1"

# 2. 将 SMILES 解析为 RDKit 的分子对象 (Mol object)
mol = Chem.MolFromSmiles(smiles)

if mol is not None:
    # 3. 方案 A：导出为高清 PNG 图片 (尺寸 600x600，背景透明)
    #Draw.MolToFile(mol, 'source_molecule.png', size=(600, 600))
    #print("成功导出高清 PNG！")
    
    # 4. 方案 B：导出为 SVG 矢量图 (学术论文极度、极度推荐！！！)
    Draw.MolToFile(mol, 'source_molecule.svg', size=(600, 600))
    print("成功导出 SVG 矢量图！")
else:
    print("SMILES 解析失败，请检查字符串是否正确。")
```
